## Homework 2, Plot SA1 and SA2 in a single plot 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file with the specified path and sheet name
file_path = r"C:\Users\boalm\320 Code BO\Sample datasets.xlsx"
sheet_name_S1 = "SA1"
# Read the data from the Excel file
data_S1 = pd.read_excel(file_path, sheet_name=sheet_name_S1)
data_S1

sheet_name_S2 = "SA2"
# Read the data from the Excel file
data_S2 = pd.read_excel(file_path, sheet_name=sheet_name_S2)
data_S2
```

```python
# # Assuming your columns are the first (strain) and second (stress)
# strain_S1 = data_S1.iloc[:, 1]  # First column (elongation)
# load_S1 = data_S1.iloc[:, 2]  # Second column (forces, can convert to stress = force/dogbone area)
# or use pandas with name
strain_S1 = data_S1['S:Position (in)']
strain_S2 = data_S2['S:Position (in)']
load_S1 = data_S1['S:Load (lbs)'] 
load_S2 = data_S2['S:Load (lbs)'] 

# Set global font properties for all text elements to Arial and size 16
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 16

# Plotting the stress-strain curve
plt.figure(figsize=(8, 6))
# plt.plot(strain_S1, load_S1, label="PLA-S1", color='b', linestyle='-', linewidth=2)
plt.plot(strain_S1, load_S1, label="PLA-S1", color='y', linestyle='--', linewidth=2, marker='*',markersize=8)
plt.plot(strain_S2, load_S2, label="PLA-S2", color='g', linestyle='-', linewidth=2, marker='*',markersize=8)
plt.xlim([-0.01, 0.18]) # define the range of x axis
plt.ylim() # define the range of y axis

# Adding labels and title with Arial font, size 16
plt.xlabel('Elongation ΔL (in)', fontsize=16)
plt.ylabel('Load F (lbs)', fontsize=16)
plt.legend()

# Adjust the ticks font size for both axes
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```
